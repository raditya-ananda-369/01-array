```
# TUGAS - SISTEM PENGELOLAAN NILAI MAHASISWA
# Struktur Data: Array (List)

import matplotlib.pyplot as plt

# 1. INPUT 10 NILAI
nilai = []

print("=== Input 10 Nilai Mahasiswa ===")
for i in range(10):
    n = float(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
    nilai.append(n)

print("\nData nilai:", nilai)


# 2. NILAI TERTINGGI & TERENDAH
print("\n=== Nilai Tertinggi & Terendah ===")
print("Nilai tertinggi:", max(nilai))
print("Nilai terendah :", min(nilai))


# 3. RATA-RATA
print("\n=== Rata-rata Nilai ===")
print("Rata-rata:", sum(nilai) / len(nilai))


# 4. JUMLAH MAHASISWA LULUS (>= 60)
lulus = 0
tidak_lulus = 0

for n in nilai:
    if n >= 60:
        lulus += 1
    else:
        tidak_lulus += 1

print("\n=== Data Kelulusan ===")
print("Lulus      :", lulus, "mahasiswa")
print("Tidak lulus:", tidak_lulus, "mahasiswa")

print("\n")
# 5. GRAFIK NILAI TERTINGGI & TERENDAH
plt.bar(range(1, 11), nilai)
plt.title("Grafik Nilai Tertinggi dan Terendah")
plt.xlabel("Mahasiswa ke-")
plt.ylabel("Nilai")
plt.show()
print("\n")

# 6. GRAFIK DATA KELULUSAN
plt.bar(["Lulus", "Tidak Lulus"], [lulus, tidak_lulus])
plt.title("Grafik Data Kelulusan Mahasiswa")
plt.xlabel("Status")
plt.ylabel("Jumlah Mahasiswa")
plt.show()

# 7. ANALISIS KOMPLEKSITAS TIAP OPERASI
print("\n=== Analisis Kompleksitas ===")
print(f"{'Operasi':<25} {'Waktu':>8}  {'Ruang':>8}")
print("-" * 45)
print(f"{'Input 10 Nilai':<25} {'O(n)':>8}  {'O(n)':>8}")
print(f"{'Cari Nilai Tertinggi':<25} {'O(n)':>8}  {'O(1)':>8}")
print(f"{'Cari Nilai Terendah':<25} {'O(n)':>8}  {'O(1)':>8}")
print(f"{'Hitung Rata-rata':<25} {'O(n)':>8}  {'O(1)':>8}")
print(f"{'Hitung Kelulusan':<25} {'O(n)':>8}  {'O(1)':>8}")
print(f"{'Grafik Nilai':<25} {'O(n)':>8}  {'O(n)':>8}")
print(f"{'Grafik Kelulusan':<25} {'O(1)':>8}  {'O(1)':>8}")
print("-" * 45)
print("\nKesimpulan:")
print("- Semua operasi utama berjalan O(n) — LINEAR")
print("- Grafik kelulusan O(1) karena hanya 2 kategori")
```
